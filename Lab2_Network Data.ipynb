```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import pandas as pd
```

## Who are the popular?

```python
data1 = pd.read_csv('dolphin_edges.csv', header=0) 
data2 = pd.read_csv('dolphin_vertices.csv', header=0)

data1
```

```python
Graph1 = nx.from_pandas_edgelist(data1 , source = 'From' ,target = 'To' ,create_using=nx.MultiDiGraph)
nx.draw(Graph1, with_labels=True, node_size=500, node_color='#FF3333')
```

```python
print(nx.info(Graph1))
```

```python
A = pd.DataFrame(Graph1.in_degree, columns=['From','To'])
A[A['To'] == max(A['To'])]
```

```python
B = pd.DataFrame(Graph1.out_degree, columns=['From','To'])
B[B['To'] == max(B['To'])]
```

```python
C = pd.DataFrame(Graph1.degree, columns=['From','To'])
C[C['To'] == max(C['To'])]
```
